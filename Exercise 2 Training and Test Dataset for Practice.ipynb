# Import the Libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
np.random.seed(2)
```

```python
x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100)/x
```

# x-axis show the count the minutes before the purchasing.

# y-axis shows the rupees of purchasing. how much money spent in purchasing.

```python
plt.scatter(x,y)
plt.show()
```

# Split the training and test data.

```python
# Training data set contain 80% of our data.
train_x = x[:80]
train_y = y[:80]
```

```python
# Test dataset contain 20% of our data.
test_x = x[80:]
test_y = y[80:]
```

# Display the Training Set.

```python
# Show the training data we use scatter plot.
plt.scatter(train_x, train_y)
plt.show()
```

# Display the Test Data.

```python
# show the test data we use scatter plot.
plt.scatter(test_x, test_y)
plt.show()
```

# Fit the dataset . we use Polnomial regression because this is best fit for data set.

```python
mymodel = np.poly1d(np.polyfit(train_x, train_y, 4))
```

```python
myline = np.linspace(0, 6, 100)
```

```python
# show the fit data using scatter plot.
plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

# We use R2 (R squared) to check the accuracy of our train-fit-model.

```python
r2 = r2_score(train_y, mymodel(train_x))
print(r2)
```

# We use R2 (R squared) to check the accuracy of our test-fit-model.

```python
r2 = r2_score(test_y, mymodel(test_x))
print(r2)
```

# Deploy our model to predict if customer stay in 5 minutes in our shop. How much money who spent? 

```python
print('Customer spent {:.2f} dollars in five minutes.'.format(mymodel(5)))
```
